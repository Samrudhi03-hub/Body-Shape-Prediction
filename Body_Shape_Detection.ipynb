```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('body_shape_dataset.csv')
data
```

```python
data['Body Shape'].unique()
```

```python
data['Body Shape'] = data['Body Shape'].replace({'Pear':1,  'Rectangle':2,  'Apple':3,  'Hourglass':4 })
```

```python
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[['Height (cm)','Weight (kg)','Bust (cm)','Waist (cm)','Hips (cm)']],data['Body Shape'], test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn import linear_model
mymodel = linear_model.LogisticRegression(max_iter=120)
mymodel.fit(X_train,y_train)
```

```python
mymodel.predict(X_test)
```

```python
X_test
```

```python
mymodel.score(X_test, y_test)
```

```python
predicted_output = mymodel.predict(X_test)
predicted_output
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_output)
cm
```

```python
import seaborn as sn
from matplotlib import pyplot as plt
plt.figure(figsize = (5,4))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted Value')
plt.ylabel('Truth or Actual Value')
```

```python
def classify_new_data(height, weight, bust, waist, hips):
    # Create a DataFrame for the new data
    new_data = pd.DataFrame({
        'Height (cm)': [height],
        'Weight (kg)': [weight],
        'Bust (cm)': [bust],
        'Waist (cm)': [waist],
        'Hips (cm)': [hips]
    })

    # Make a prediction
    prediction = mymodel.predict(new_data)

    # Map the prediction back to the original labels
    body_shape_labels = {1: 'Pear', 2: 'Rectangle', 3: 'Apple', 4: 'Hourglass'}
    predicted_body_shape = body_shape_labels[prediction[0]]

    return predicted_body_shape

```

```python
# Example usage
new_height = 170
new_weight = 68
new_bust = 90
new_waist = 80
new_hips = 85

```

```python
predicted_body_shape = classify_new_data(new_height, new_weight, new_bust, new_waist, new_hips)
print(f'The predicted body shape is: {predicted_body_shape}')
```
